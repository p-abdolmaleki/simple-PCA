```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PCA import PCA
```

```python
np.random.seed(42)
temp = 9 * np.random.rand(500) + 1
df = pd.DataFrame({'x1':temp, 'x2':temp + np.random.rand(500) - 0.5})
df['label'] = np.where((df['x1'] + df['x2']) > 11, 1, 0)
plt.scatter(df['x1'], df['x2'], c=df['label'])
plt.show()
```

```python
pca = PCA(n_components=2)
new_df = df.copy()
new_df.iloc[:, :2] = pca.fit_transform(df.iloc[:, :2])
plt.scatter(new_df['x1'], new_df['x2'], c=new_df['label'])
plt.ylim(-3, 3)
plt.xlim(-3, 3)
plt.show()
```
